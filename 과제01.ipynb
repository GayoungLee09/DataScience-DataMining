```python
#c  Show the histogram of price! Select the proper bins! Explain!
plt.hist(data['price'], bins = 10) #histogram with bins (class interval) = 10 (equals part)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
#d  Analyze the “depth” and “table”! What is the best chart to answer this problem? Explain the relationship between “depth” and “table”, if any!
plt.scatter(data['depth'], data['table'])
plt.title("depth vs table")
plt.xlabel("depth")
plt.ylabel("table")
plt.show()
```

The data are gathered at values between 55 and 70, and tables between 50 and 70.

```python
#e  Using Faceted Histogram, show the histogram of the carat with respect to the 'cut' and 'color'! Explain!
import seaborn as sns
sns.set()
grid = sns.FacetGrid(data, row="cut", col="color", margin_titles=True)
grid.map(plt.hist, "carat");
```

When the color = J, cut = Good, cut = Fair, the values of the frequency distribution table are smaller than the other values.

```python
#f Using Factor Plots, show the boxplot of the carat with respect to the 'cut' and 'color'! Explain!
with sns.axes_style(style='ticks'):
    g = sns.catplot("color","carat", "cut", data=data, kind="box")
    g.set_axis_labels("color", "carat");
```

In all colors, the carat value of the fair cut is higher than the carat value of the same color.

```python
tips = sns.load_dataset('tips')
```

```python
#a Show the heatmap for the “tips” dataset! What is your argument based on the result?
sns.heatmap(tips.corr(), annot=True)
```

The tip and total_bill have the highest correlation and the tip and size have the lowest correlation.

```python
#b Show the pairplot for the “tips” dataset! (set the hue = day!). What is your argument based on the result?
sns.pairplot(tips, hue='day', diag_kws={'bw':'1.0'}); 
```

We can see that people tip more on Saturdays and Sundays than on weekdays.

```python
#c Using Factor Plots, show the boxplot of the tips percentage with respect to the 'day' and 'time'! Explain!
tips['tip_percent'] = 100 * tips['tip'] / tips['total_bill']
with sns.axes_style(style='ticks'):
    g = sns.catplot("day","tip_percent", "time", data=tips, kind="box")
    g.set_axis_labels("day", "tip_percent");
```

We can see that tip happened on dinner time only on weekends.

```python
#d  Using Faceted Histogram, show the histogram of the tips percentage with respect to the 'time' and 'size'! Explain
grid = sns.FacetGrid(tips, row="time", col="size", margin_titles=True)
grid.map(plt.hist, "tip_percent");
```

When the size = 2, time = Dinner, the values of the frequency distribution table are bigger than the other values.
